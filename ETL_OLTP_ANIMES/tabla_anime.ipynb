```python
import pandas as pd
import ast
pd.set_option('display.max_columns', None)
```

```python
anime = df_aniime[["anime_id","name","type","episodes","image_url",'aired',"duration",'score',"scored_by",'rank','popularity','favorites','Synopsis','status','source','rating_y']]
```

```python
# Convertir las cadenas en la columna 'aired' en diccionarios reales
anime['aired'] = anime['aired'].apply(ast.literal_eval)
```

```python
# Crear una función para aplicar a la columna 'aired'
def extract_dates(aired):
    if aired['from'] is None:
        start_date = None
    else:
        start_date = aired['from']
    if aired['to'] is None:
        finish_date = None
    else:
        finish_date = aired['to']
    return pd.Series({'START_DATE': start_date, 'FINISH_DATE': finish_date})
```

```python
# Aplicar la función a la columna 'aired' para crear nuevas columnas 'START_DATE' y 'FINISH_DATE'
anime[['START_DATE', 'FINISH_DATE']] = anime['aired'].apply(extract_dates)

# Eliminar la columna 'aired' si ya no la necesitas
anime.drop(columns=['aired'], inplace=True)

anime
```

```python
df_clasificacion = pd.read_csv('./TABLAS/CLASIFICATION.csv')
df_clasificacion
```

```python
df_source = pd.read_csv('./TABLAS/SOURCE.csv')
df_source
```

```python
# Unir DataFrame 'anime' con 'df_clasificacion' para obtener los IDs de clasificación
anime = anime.merge(df_clasificacion, how='left', left_on='rating_y', right_on='CLASSIFICATION')

# Rellenar valores nulos en 'id' con -1
anime['ID_CLASSIFICATION'] = anime['ID_CLASSIFICATION'].fillna(-1)
```

```python
# Eliminar la columna 'clasificacion' si ya no la necesitas
anime.drop(columns=['CLASSIFICATION','rating_y'], inplace=True)
```

```python
anime
```

```python
# Unir DataFrame 'anime' con 'df_source' para obtener los IDs de source
anime = anime.merge(df_source, how='left', left_on='source', right_on='SOURCE')
```

```python
# Eliminar la columna 'SOURCE' si ya no la necesitas
anime.drop(columns=['SOURCE','source'], inplace=True)
```

```python
anime.columns
```

```python
df_type = pd.read_csv("./TABLAS/TYPES.csv")
df_type
```

```python
# Unir DataFrame 'anime' con 'df_clasificacion' para obtener los IDs de clasificación
anime = anime.merge(df_type, how='left', left_on='type', right_on='TYPE')

# Rellenar valores nulos en 'id' con -1
anime['ID_TYPE'] = anime['ID_TYPE'].fillna(-1)

# Eliminar la columna 'SOURCE' si ya no la necesitas
anime.drop(columns=['type','TYPE'], inplace=True)
anime
```

```python
# Crear un diccionario para mapear los nombres de las columnas antiguas a las nuevas
nuevos_nombres = {
    'anime_id': 'ANIME_ID',
    'type': 'TYPE',
    'name': 'ANIME_NAME',
    'episodes': 'EPISODES',
    'image_url': 'URL_IMAGE',
    'duration': 'DURATION',
    'score': 'SCORED',
    'scored_by': 'SCORED_BY',
    'rank': 'RANK',
    'popularity': 'POPULARITY',
    'favorites': 'FAVORITES',
    'Synopsis': 'SYNOPSIS',
    'status': 'STATUS',
    'START_DATE': 'START_DATE',
    'FINISH_DATE': 'FINISH_DATE',
    'ID_CLASSIFICATION': 'ID_CLASSIFICATION',
    'ID_SOURCE': 'ID_SOURCE'
}

# Renombrar las columnas del DataFrame
anime = anime.rename(columns=nuevos_nombres)
```

```python
anime
```

```python
# Convertir las columnas de identificación (IDs) a tipo int
ids_a_convertir = ['ANIME_ID', 'ID_CLASSIFICATION', 'ID_SOURCE','ID_TYPE']
anime[ids_a_convertir] = anime[ids_a_convertir].astype(int)
```

```python
anime.to_csv('./TABLAS/ANIME.csv',index=False)
```

```python
import pandas as pd
```

```python
df_anime = pd.read_csv('/content/ANIME (3).csv')
df_anime
```

```python
df_rating = pd.read_csv('/content/rating.csv')
df_rating
```

```python
df_hechos = pd.merge(df_rating,df_anime, left_on='anime_id', right_on='ANIME_ID',how="inner" )
df_hechos
```

```python
save_tables = df_rating.columns.to_list()
save_tables += ["ID_CLASSIFICATION","ID_SOURCE","ID_TYPE"]
```

```python
df_hechos = df_hechos[list(set(save_tables).intersection(df_hechos.columns))]
df_hechos
```

```python
ordeando = ['anime_id', 'ID_CLASSIFICATION', 'ID_SOURCE', 'user_id', 'ID_TYPE', 'rating']
```

```python
df_organizado = df_hechos.reindex(columns=ordeando)
df_organizado
```

```python
renombrar = {
    'anime_id': 'ID_ANIME',
    'ID_CLASSIFICATION': 'ID_CLASSIFICATION',
    'ID_SOURCE': 'ID_SOURCE',
    'user_id': 'ID_USER',
    'ID_TYPE': 'ID_TYPES',
    'rating': 'RATING'
}
```

```python
df_Hechos_final = df_organizado.rename(columns=renombrar)
df_Hechos_final
```

```python
df_Hechos_final.to_csv('./ANIME_HECHOS.csv')
```

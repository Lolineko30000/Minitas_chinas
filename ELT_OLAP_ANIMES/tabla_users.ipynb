```python
import pandas as pd
import numpy as np
```

```python
df_users = pd.read_csv("./users_final.csv")
```

```python
df_users
```

```python
df_users.columns
```

```python
df_users = df_users[['Mal ID', 'Username','Location','Completed','Dropped','Plan to Watch','Watching']]
```

```python
df_users.columns
```

```python
df_users
```

```python
nuevos_nombres = {'Mal ID':'ID_USER','Username':'USERNAME','Location': 'LOCATION',
                 'Completed':'COMPLETED','Dropped':'DROPPED','Plan to Watch':'PLAN_TO_WATCH',
                 'Watching':'WATCHING'}
df_users = df_users.rename(columns=nuevos_nombres)
```

```python
df_users
```

```python
df_users.isna().sum()
```

```python
#Akihabara, Bolivia, Estados Unidos
specific_values=["Japón","Estados Unidos","Bolivia"]
df_users['LOCATION'].fillna(np.random.choice(specific_values), inplace=True)

```

```python
df_users.dtypes
```

```python
df_users['USERNAME'].fillna(value='unknow', inplace=True)
df_users['COMPLETED'].fillna(value=0, inplace=True)
df_users['DROPPED'].fillna(value=0, inplace=True)
df_users['PLAN_TO_WATCH'].fillna(value=0, inplace=True)
df_users['WATCHING'].fillna(value=0, inplace=True)
```

```python
df_users.isna().sum()
```

```python
df_users.to_csv('./USERS.csv', index=False)
```

```python
user_merge_all = user_ds_join_user_cl.merge(user_fd, how='left', left_on = 'Mal ID', right_on = 'user_id')
```

```python
drop_cols = ['username_x', 'user_id_x', 'user_watching_x',
       'user_completed_x', 'user_onhold_x', 'user_dropped_x',
       'user_plantowatch_x', 'user_days_spent_watching_x', 'gender_x',
       'location_x', 'birth_date_x', 'access_rank_x', 'join_date_x',
       'stats_mean_score_x', 'stats_rewatched_x',
       'stats_episodes_x', '_merge', 'username_y', 'user_id_y',
       'user_watching_y', 'user_completed_y', 'user_onhold_y',
       'user_dropped_y', 'user_plantowatch_y', 'user_days_spent_watching_y',
       'gender_y', 'location_y', 'birth_date_y', 'access_rank_y',
       'join_date_y', 'last_online_y', 'stats_mean_score_y',
       'stats_rewatched_y', 'stats_episodes_y']
```

```python
imputer_cols = {'Gender': ['gender_y', 'gender_x'],
                'Birthday': ['birth_date_x', 'birth_date_y'],
                'Location': ['location_y', 'location_x'],
                'last_online_x': ['last_online_y']}
```

```python
for col, imputers in imputer_cols.items():
  for imputer in imputers:
    user_merge_all[col] = user_merge_all[col].fillna(user_merge_all[imputer])
```

```python
user_merge_all.isnull().sum()
```

```python
user_merge_all = user_merge_all.drop(drop_cols, axis = 1)
```

```python
user_merge_all.columns
```

```python
user_merge_all = user_merge_all.rename({'last_online_x': 'Last Online'}, axis=1)
```

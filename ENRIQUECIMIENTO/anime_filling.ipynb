```python
mal_anime_drop = ['title','title_english' ,'title_japanese','title_synonyms',
                  'background', 'premiered', 'broadcast', 'licensor', 'producer',
                  'genre', 'type', 'episodes', 'members']
```

```python
anime_join_mal_anime = anime.merge(mal_anime, how = 'inner', on='anime_id')
```

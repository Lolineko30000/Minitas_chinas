```python
# Ordenar los animes por anime_id
animes_ordenados_por_id = anime.sort_values(by='anime_id')
```

```python
animes_ordenados_por_id.head()
```

```python
# Obtener los conjuntos de anime_id de cada DataFrame
anime_id_animes_ordenados = set(animes_ordenados_por_id['anime_id'])
anime_id_anime_2023 = set(anime_2023['anime_id'])
```

```python
# Encontrar la intersección de los conjuntos
anime_id_comunes = anime_id_animes_ordenados.intersection(anime_id_anime_2023)
```

```python
# Imprimir los anime_id comunes
print(anime_id_comunes)
```

```python
print(len(anime_id_comunes))
```

```python
# Filtrar ambos conjuntos de datos con los anime_id comunes
animes_ordenados_comunes = animes_ordenados_por_id[animes_ordenados_por_id['anime_id'].isin(anime_id_comunes)]
anime_2023_comunes = anime_2023[anime_2023['anime_id'].isin(anime_id_comunes)]
```

```python
# Verificar si los nombres coinciden
for anime_id in anime_id_comunes:
    nombre_anime_ordenados = animes_ordenados_comunes[animes_ordenados_comunes['anime_id'] == anime_id]['name'].iloc[0]
    nombre_anime_2023 = anime_2023_comunes[anime_2023_comunes['anime_id'] == anime_id]['Name'].iloc[0]
    if nombre_anime_ordenados != nombre_anime_2023:
        print(f"¡Advertencia! El anime con ID {anime_id} tiene nombres diferentes:")
        print(f"Nombre en animes_ordenados_por_id: {nombre_anime_ordenados}")
        print(f"Nombre en anime_2023: {nombre_anime_2023}")
```

```python
# Obtener los anime_id de cada DataFrame como conjuntos
anime_id_animes_ordenados = set(animes_ordenados_por_id['anime_id'])
anime_id_anime_2023 = set(anime_2023['anime_id'])
```

```python
# Encontrar los anime_id que están en animes_ordenados_por_id pero no en anime_2023
anime_id_faltantes = anime_id_animes_ordenados.difference(anime_id_anime_2023)
```

```python
# Filtrar el DataFrame animes_ordenados_por_id para obtener los nombres de los animes faltantes
animes_ordenados_por_id[animes_ordenados_por_id['anime_id'].isin(anime_id_faltantes)][['anime_id', 'name']]
```

```python
# Obtener los anime_id que están en anime pero no en anime_2023
anime_id_faltantes = set(anime['anime_id']).difference(set(anime_2023['anime_id']))
```

```python
# Filtrar anime para obtener los datos de los animes faltantes
animes_diferentes_2023 = anime[anime['anime_id'].isin(anime_id_faltantes)]
```

Unir los Data set de anime

```python
# Asegúrate de que las columnas tengan los mismos nombres
animes_diferentes_2023.rename(columns={'name': 'Name','genre': 'Genres', 'type' : 'Type', 'episodes': 'Episodes', 'rating': 'Rank', 'members': 'Members'}, inplace=True)
```

```python
# Completa las columnas faltantes en animes_diferentes_2023
columnas_faltantes = ['English name', 'Other name', 'Score', 'Synopsis', 'Aired', 'Premiered', 'Status', 'Producers', 'Licensors', 'Studios', 'Source', 'Duration', 'Rating', 'Popularity', 'Favorites', 'Scored By', 'Image URL']
for columna in columnas_faltantes:
    if columna not in animes_diferentes_2023.columns:
        animes_diferentes_2023[columna] = ''
```

```python
# Obtener los user_id coincidentes
user_ids_coincidentes = rating_animes_diferentes_2023[rating_animes_diferentes_2023['user_id'].isin(user_det['Mal ID'])]['user_id'].unique()
```

```python
# Obtener los user_id no coincidentes
user_ids_no_coincidentes = rating_animes_diferentes_2023[~rating_animes_diferentes_2023['user_id'].isin(user_det['Mal ID'])]['user_id'].unique()
```

```python
user_ids_no_coincidentes
```

```python
# Crear un DataFrame con los user_ids no coincidentes
df_no_coincidentes = pd.DataFrame({'Mal ID': user_ids_no_coincidentes})
```

```python
# Llenar los valores faltantes con 'UNKNOWN' para las columnas de tipo objeto y 'NULL' para las columnas de tipo float64
df_no_coincidentes = df_no_coincidentes.fillna({'Gender': 'UNKNOWN', 'Birthday': 'UNKNOWN', 'Location': 'UNKNOWN'})
df_no_coincidentes = df_no_coincidentes.fillna(0)
```

```python
# Concatenar el DataFrame de user_ids no coincidentes con user_det
user_det = pd.concat([user_det, df_no_coincidentes], ignore_index=True)
```
